## Objective

Transform the raw EM-DAT disaster dataset into a standardized analytical dataset containing cleaned historical flood events for India.


```python
# ============================================================
# Notebook 03 : Silver Exposure Layer
# India Flood Risk Platform
# ============================================================

# Standard Library
from pathlib import Path

# Data Handling
import pandas as pd
import numpy as np

# Display
pd.set_option("display.max_columns", None)
pd.set_option("display.max_rows", 10)

print("Libraries loaded successfully!")
```

```python
# ============================================================
# Project Paths
# ============================================================

PROJECT_ROOT = Path.cwd().parent

EMDAT_FILE = (
    PROJECT_ROOT
    / "data"
    / "raw"
    / "emdat"
    / "EMDAT_India_Floods_2000_2025.xlsx"
)

SILVER_DIR = (
    PROJECT_ROOT
    / "data"
    / "silver"
)

print("Project Root :", PROJECT_ROOT)
print("File Exists :", EMDAT_FILE.exists())
```

```python
# ============================================================
# Load EM-DAT Dataset
# ============================================================

df = pd.read_excel(EMDAT_FILE)

print("Dataset loaded successfully!")
```

```python
# Shape

print(df.shape)

df.head()
```

```python
df.info()
```

```python
df.columns.tolist()
```

```python
# ============================================================
# Select Analytical Columns
# ============================================================

selected_columns = [

    # Identification
    "DisNo.",
    "Historic",

    # Disaster Classification
    "Disaster Type",
    "Disaster Subtype",

    # Geography
    "Country",
    "Location",
    "Latitude",
    "Longitude",
    "River Basin",

    # Event Dates
    "Start Year",
    "Start Month",
    "Start Day",
    "End Year",
    "End Month",
    "End Day",

    # Human Impact
    "Total Deaths",
    "No. Affected",
    "Total Affected",

    # Economic Impact
    "Total Damage ('000 US$)",
    "Total Damage, Adjusted ('000 US$)",

    # Administrative Units
    "Admin Units"
]

emdat = df[selected_columns].copy()

emdat.head()
```

```python
print(emdat.shape)

emdat.info()
```

```python
# ============================================================
# Rename Columns
# ============================================================

emdat.rename(
    columns={
        "DisNo.": "event_id",
        "Historic": "historic",
        "Disaster Type": "disaster_type",
        "Disaster Subtype": "disaster_subtype",
        "Country": "country",
        "Location": "location",
        "Latitude": "latitude",
        "Longitude": "longitude",
        "River Basin": "river_basin",
        "Start Year": "start_year",
        "Start Month": "start_month",
        "Start Day": "start_day",
        "End Year": "end_year",
        "End Month": "end_month",
        "End Day": "end_day",
        "Total Deaths": "total_deaths",
        "No. Affected": "people_affected",
        "Total Affected": "total_affected",
        "Total Damage ('000 US$)": "damage_usd_000",
        "Total Damage, Adjusted ('000 US$)": "damage_adjusted_usd_000",
        "Admin Units": "admin_units"
    },
    inplace=True
)

emdat.head()
```

```python
# ============================================================
# Missing Value Analysis
# ============================================================

missing = pd.DataFrame({
    "Missing Values": emdat.isna().sum(),
    "Percentage": (
        emdat.isna().mean() * 100
    ).round(2)
})

missing.sort_values(
    "Percentage",
    ascending=False
)
```

```python
# ============================================================
# Data Types
# ============================================================

emdat.dtypes
```

```python
# ============================================================
# Fill Missing Human Impact Values
# ============================================================

impact_columns = [
    "total_deaths",
    "people_affected",
    "total_affected"
]

emdat[impact_columns] = (
    emdat[impact_columns]
    .fillna(0)
)

emdat[impact_columns].isna().sum()
```

```python
# ============================================================
# Create Event Dates
# ============================================================

emdat["start_date"] = pd.to_datetime(
    dict(
        year=emdat["start_year"],
        month=emdat["start_month"],
        day=emdat["start_day"].fillna(1)
    ),
    errors="coerce"
)

emdat["end_date"] = pd.to_datetime(
    dict(
        year=emdat["end_year"],
        month=emdat["end_month"],
        day=emdat["end_day"].fillna(1)
    ),
    errors="coerce"
)

emdat[
    [
        "start_date",
        "end_date"
    ]
].head()
```

```python
# ============================================================
# Event Duration
# ============================================================

emdat["duration_days"] = (
    emdat["end_date"] -
    emdat["start_date"]
).dt.days + 1

emdat[
    [
        "event_id",
        "start_date",
        "end_date",
        "duration_days"
    ]
].head()
```

```python
# ============================================================
# Organize Final Dataset
# ============================================================

emdat = emdat[
    [

        # Event
        "event_id",
        "historic",

        # Classification
        "disaster_type",
        "disaster_subtype",

        # Geography
        "country",
        "location",
        "river_basin",
        "latitude",
        "longitude",
        "admin_units",

        # Dates
        "start_date",
        "end_date",
        "duration_days",

        # Original Date Components
        "start_year",
        "start_month",
        "start_day",
        "end_year",
        "end_month",
        "end_day",

        # Human Impact
        "total_deaths",
        "people_affected",
        "total_affected",

        # Economic Impact
        "damage_usd_000",
        "damage_adjusted_usd_000"
    ]
]

emdat.head()
```

```python
# ============================================================
# Validation
# ============================================================

print("=" * 60)
print("EM-DAT SILVER VALIDATION")
print("=" * 60)

print("Rows :", len(emdat))
print("Columns :", len(emdat.columns))

print("\nDate Range")
print(emdat["start_date"].min())
print(emdat["end_date"].max())

print("\nMissing Start Dates")
print(emdat["start_date"].isna().sum())

print("\nDuplicate Event IDs")
print(emdat["event_id"].duplicated().sum())
```

```python
# ============================================================
# Export Silver Dataset
# ============================================================

silver_file = (
    SILVER_DIR /
    "silver_emdat_events.parquet"
)

emdat.to_parquet(
    silver_file,
    index=False
)

print("Silver EM-DAT exported successfully!")
print(silver_file)
```

```python
# ============================================================
# Engineering Summary
# ============================================================

print("=" * 60)
print("SILVER EXPOSURE LAYER")
print("=" * 60)

print(f"Flood Events           : {len(emdat):,}")
print(f"Variables              : {len(emdat.columns)}")
print(f"Countries              : {emdat['country'].nunique()}")
print(f"Date Range             : {emdat['start_year'].min()} - {emdat['end_year'].max()}")

print("\nOutputs")
print("----------------------")
print("silver_emdat_events.parquet")
```
