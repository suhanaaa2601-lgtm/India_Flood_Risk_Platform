## Bronze Layer Verification

### Objective

Verify that the exported Bronze Layer Parquet file has been created correctly and validate its structure before it is used in downstream processing.

```python
from pathlib import Path
import pandas as pd

PROJECT_ROOT = Path.cwd().parent

BRONZE_FILE = (
    PROJECT_ROOT
    / "data"
    / "bronze"
    / "bronze_rainfall_2023.parquet"
)

print(BRONZE_FILE)
print(BRONZE_FILE.exists())
```

```python
bronze = pd.read_parquet(BRONZE_FILE)

bronze.head()
```

```python
bronze.info()
```

```python
bronze.describe()
```

```python
print("="*60)
print("BRONZE DATASET VALIDATION")
print("="*60)

print(f"Rows: {bronze.shape[0]:,}")
print(f"Columns: {bronze.shape[1]}")
print(f"Date Range: {bronze['date'].min()} to {bronze['date'].max()}")
print(f"Missing Values: {bronze.isna().sum().sum()}")
print(f"Duplicate Rows: {bronze.duplicated().sum()}")

print("\nValidation Complete")
```
